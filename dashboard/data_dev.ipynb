```python
mu = []
def solve_metions(x):
    for i, ms in enumerate(x):
        ous = []
        mtsasts = ast.literal_eval(ms)
        mtsast = [x for x in mtsasts]
        if len(mtsast) > 0:
            for l in mtsast:
                ou = l['screenName']
                ous.append(ou)
        mu.append(ous)
```

```python
dw.data[['mentions']].apply(solve_metions)
dw.data['mentions.screennames'] = mu
```
